### Uniform Distribution

```python

def UD(a,b,x):
    y1 = (1/(b-a))*(x/x)
    y2 = ((x - a) / (b - a))
    y3 = ((a+b)/2)*(x/x)
    y4 = ((a+b)/2)*(x/x)
    y5 = (((b-a)**2)/12)*(x/x)
    z = (y1,y2,y3,y4,y5)
    return z

(a,b) = (3,5)
(c,d) = (2,6)
(e,f) = (1,9)

x1 = np.linspace(a,b,100)
x2 = np.linspace(c,d,100)
x3 = np.linspace(e,f,100)

UD1 = UD(a,b,x1)
UD2 = UD(c,d,x2)
UD3 = UD(e,f,x3)

fig, ax = plt.subplots(1,3,figsize=(15,15))
ax[0].plot(x1,UD1[0], ls='-')
ax[0].plot(x1,UD1[1], ls='-')
ax[0].plot(x1,UD1[2], ls=':')
ax[0].plot(x1,UD1[3], ls='-.')
ax[0].plot(x1,UD1[4], ls='--')
ax[0].set_title("Uniform Distribution \n first variable range")
ax[0].set_xlabel("X")
ax[0].set_ylabel("Y")
ax[0].legend(['PDF','CDF','MEAN','MEDIAN','VARIANCE'], prop={'size':5})


ax[1].plot(x2,UD2[0], ls='-')
ax[1].plot(x2,UD2[1], ls='-')
ax[1].plot(x2,UD2[2], ls=':')
ax[1].plot(x2,UD2[3], ls='-.')
ax[1].plot(x2,UD2[4], ls='--')
ax[1].set_title("Uniform Distribution \n Second variable range ")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Y")
ax[1].legend(['PDF','CDF','MEAN','MEDIAN','VARIANCE'], prop={'size':5})

ax[2].plot(x3,UD3[0], ls='-')
ax[2].plot(x3,UD3[1], ls='-')
ax[2].plot(x3,UD3[2], ls=':')
ax[2].plot(x3,UD3[3], ls='-.')
ax[2].plot(x3,UD3[4], ls='--')
ax[2].set_title("Uniform Distribution \n Third variable range")
ax[2].set_xlabel("X")
ax[2].set_ylabel("Y")
ax[2].legend(['PDF','CDF','MEAN','MEDIAN','VARIANCE'], prop={'size':5})

plt.show()
plt.tight_layout()
```

### Exponential Distribution

```python
def ED(a,x):
    e = math.exp(1)
    y1 = (a*(e**(-a*x)))*(x/x)
    y2 = (1 - (e ** (-a * x))) * (x / x)
    y3 = (1 / a) * (x / x)
    l = math.log(2, e)
    y4 = (l / a) * (x / x)
    y5 = (1 / (a ** 2)) * (x / x)
    z = (y1, y2, y3, y4, y5)
    return z


x1_ED = np.linspace(1,10,1000)
x2_ED = np.linspace(1,80,1000)
x3_ED = np.linspace(1,200,1000)

f = 0.5
g = 1
h = 2

ED1 = ED(f,x1_ED)
ED2 = ED(g,x2_ED)
ED3 = ED(h,x3_ED)


fig_ED, ax_ED = plt.subplots(1,3,figsize=(15,15))

ax_ED[0].plot(x1_ED,ED1[0], ls='-')
ax_ED[0].plot(x1_ED,ED1[1], ls='-')
ax_ED[0].plot(x1_ED,ED1[2], ls=':')
ax_ED[0].plot(x1_ED,ED1[3], ls='-.')
ax_ED[0].plot(x1_ED,ED1[4], ls='--')
ax_ED[0].set_title("Exponetial Distribution \n first variable range")
ax_ED[0].set_xlabel("X")
ax_ED[0].set_ylabel("Y")
ax_ED[0].legend(['PDF','CDF','MEAN','MEDIAN','VARIANCE'], prop={'size':5})


ax_ED[1].plot(x2_ED,ED2[0], ls='-')
ax_ED[1].plot(x2_ED,ED2[1], ls='-')
ax_ED[1].plot(x2_ED,ED2[2], ls=':')
ax_ED[1].plot(x2_ED,ED2[3], ls='-.')
ax_ED[1].plot(x2_ED,ED2[4], ls='--')
ax_ED[1].set_title("Exponetial Distribution \n Second variable range ")
ax_ED[1].set_xlabel("X")
ax_ED[1].set_ylabel("Y")
ax_ED[1].legend(['PDF','CDF','MEAN','MEDIAN','VARIANCE'], prop={'size':5})

ax_ED[2].plot(x3_ED,ED3[0], ls='-')
ax_ED[2].plot(x3_ED,ED3[1], ls='-')
ax_ED[2].plot(x3_ED,ED3[2], ls=':')
ax_ED[2].plot(x3_ED,ED3[3], ls='-.')
ax_ED[2].plot(x3_ED,ED3[4], ls='--')
ax_ED[2].set_title("Exponetial Distribution \n Third variable range")
ax_ED[2].set_xlabel("X")
ax_ED[2].set_ylabel("Y")
ax_ED[2].legend(['PDF','CDF','MEAN','MEDIAN','VARIANCE'], prop={'size':5})



plt.show()
plt.tight_layout()
```

### Gamma Distribution

```python
def gamma(a):
    b=1
    for i in range(1,a):
        b= i*b
    return b


def GD(a, b, x):
    e = math.exp(1)
    y1 = ((b**a)*(x**(a-1))*(e**(-b*x)))/(gamma(a))*(x/x)
    y2 = gammainc(a, x*b)*(x/x)
    y3 = (a/b)*(x/x)
    y4 = (a-1/b)*(x/x)
    y5 = (a/(b**2))*(x/x)
    z = (y1, y2, y3, y4, y5)
    return z

x1_GD = np.linspace(1,12,100)
x2_GD = np.linspace(1,12,100)
x3_GD = np.linspace(1,12,100)

(i,j) = (5,4)
(k,l) = (7,1)
(m,n) = (9,3)

GD1 = GD(i,j,x1_GD)
GD2 = GD(k,l,x2_GD)
GD3 = GD(m,n,x3_GD)

fig_GD, ax_GD = plt.subplots(1,3,figsize=(15,15))

ax_GD[0].plot(x1_GD,GD1[0], ls='-')
ax_GD[0].plot(x1_GD,GD1[1], ls='-')
ax_GD[0].plot(x1_GD,GD1[2], ls=':')
ax_GD[0].plot(x1_GD,GD1[3], ls='-.')
ax_GD[0].plot(x1_GD,GD1[4], ls='--')
ax_GD[0].set_title("GAMMA Distribution \n first variable range")
ax_GD[0].set_xlabel("X")
ax_GD[0].set_ylabel("Y")
ax_GD[0].legend(['PDF','CDF','MEAN','MODE','VARIANCE'], prop={'size':5})


ax_GD[1].plot(x2_GD,GD2[0], ls='-')
ax_GD[1].plot(x2_GD,GD2[1], ls='-')
ax_GD[1].plot(x2_GD,GD2[2], ls=':')
ax_GD[1].plot(x2_GD,GD2[3], ls='-.')
ax_GD[1].plot(x2_GD,GD2[4], ls='--')
ax_GD[1].set_title("GAMMA Distribution \n Second variable range ")
ax_GD[1].set_xlabel("X")
ax_GD[1].set_ylabel("Y")
ax_GD[1].legend(['PDF','CDF','MEAN','MODE','VARIANCE'], prop={'size':5})

ax_GD[2].plot(x3_GD,GD3[0], ls='-')
ax_GD[2].plot(x3_GD,GD3[1], ls='-')
ax_GD[2].plot(x3_GD,GD3[2], ls=':')
ax_GD[2].plot(x3_GD,GD3[3], ls='-.')
ax_GD[2].plot(x3_GD,GD3[4], ls='--')
ax_GD[2].set_title("GAMMA Distribution \n Third variable range")
ax_GD[2].set_xlabel("X")
ax_GD[2].set_ylabel("Y")
ax_GD[2].legend(['PDF','CDF','MEAN','MODE','VARIANCE'], prop={'size':5})

plt.show()
plt.tight_layout()
```

### b) Binomal Distribution

```python
def Binomal(n, r, p):

    y1 = ((Combination(n,r))*(p**r)*((1-p)**(n-r)))
    return y1

def Combination(n, r):
    x = (math.factorial(n))/((math.factorial(r))*(math.factorial(n-r)))
    return x

success2 = np.linspace(0,20,21)
p_b_1 = 0.4

N_b = 20
list_b = np.zeros(21)
for i in range(N_b+1):
    list_b[i] = Binomal(N_b,i,p_b_1)
mean_b = N_b*p_b_1*((success2+1)/(success2+1))
variance_b = N_b*p_b_1*(1-p_b_1)*((success2+1)/(success2+1))

fig_b, ax_b = plt.subplots(1,2,figsize=(15,8))
ax_b[0].set_xlim(-0.2,22)
ax_b[0].set_ylim(-0.01,0.5)
ax_b[0].scatter(success2, list_b, color='black', label='pmf')
ax_b[0].bar(success2, list_b,color='black',width=0.05)
ax_b[1].plot(success2, mean_b,'--g', label = 'mean_b')
ax_b[1].plot(success2, variance_b,':r', label = 'variance_b')
ax_b[0].set_title("PMF")
ax_b[1].set_title("MEAN & VARIANCE")
ax_b[0].legend()
ax_b[1].legend()
plt.tight_layout()
```

### d) Poisson Distribution

```python
def fact(a):
    b=1
    for i in range(1,a+1):
        b= i*b
    return b

def Poisson(a, x):

    y1 = ((np.exp(-a))*(a**x))/(fact(x))
    return y1


success4 = np.linspace(1,10,10)
occurances = 5

list_p = np.zeros(10)
for i_p in range(10):
    list_p[i_p] = Poisson(occurances,i_p)
mean_p = (occurances)*((success4+1)/(success4+1))
variance_p =(occurances)*((success4+1)/(success4+1))

fig_p, ax_p = plt.subplots(1,2,figsize=(15,8))
ax_p[0].set_xlim(-0.2,12)
ax_p[0].set_ylim(-0.01,0.25)
ax_p[0].scatter(success4, list_p, color='black', label='pmf')
ax_p[0].bar(success4, list_p,color='black',width=0.1)
ax_p[1].plot(success4, mean_p,'--g', label = 'mean')
ax_p[1].plot(success4, variance_p,':r', label = 'variance')
ax_p[0].set_title("PMF")
ax_p[1].set_title("MEAN & VARIANCE")
ax_p[0].legend()
ax_p[1].legend()
```
